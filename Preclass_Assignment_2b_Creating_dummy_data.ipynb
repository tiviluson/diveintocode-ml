```
import numpy as np
import matplotlib.pyplot as plt
```

# [Problem 1] Creating random numbers

```
mean = [-3, 0]
cov = [[1, 0.8], [0.8, 1]]
np.random.seed(0)
data = np.random.multivariate_normal(mean, cov, 500)
```

# [Problem 2] Visualization by scatter plot

```
plt.scatter(data[:,0], data[:,1])
plt.show()
```

# [Problem 3] Visualization by histogram

```
plt.xlim(-6,3)
plt.hist(data[:,0],bins=50)
plt.show()
plt.xlim(-6,3)
plt.hist(data[:,1],bins=50)
plt.show()
```

# [Problem 4] Addition of data

```
mean = [0, -3]
cov = [[1, 0.8], [0.8, 1]]
another_data = np.random.multivariate_normal(mean, cov, 500)
plt.clf()
plt.scatter(data[:,0], data[:,1],label="Data 1")
plt.scatter(another_data[:,0], another_data[:,1],label="Data 2")
plt.legend(loc="upper left")
plt.show()
```

# [Problem 5] Data combination

```
print(np.vstack([data, another_data]).shape)
print(np.concatenate((data, another_data)).shape)
combined_data = np.concatenate((data, another_data))
```

# [Problem 6] Labeling

```
label = np.vstack([np.zeros((500,1)), np.ones((500,1))])
labeled_data = np.hstack((combined_data, label))
print(labeled_data)
```
