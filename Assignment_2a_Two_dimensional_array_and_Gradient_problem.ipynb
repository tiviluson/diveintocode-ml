```python
import numpy as np
import matplotlib.pyplot as plt
```

# [Problem 1] Linear function

```python
x = np.arange(-50,50.1,0.1).reshape(-1,1)
y = 1/2*x + 1
print(x.shape)
print(y)
```

# [Problem 2] Array combination

```python
z = np.hstack((x,y))
print(z)
print(z.shape)
```

# [Problem 3] Find the gradient

```python
grad = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
print(grad.shape)
print(grad)
```

# [Problem 4] Draw a graph

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y(x)")
plt.show()
plt.clf
plt.plot(x[1:], grad)
plt.xlabel("x")
plt.ylabel("dy/dx")
plt.title("Gradient dy/dx")
plt.show()
```

# [Problem 5] Python functionalization

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    array_xy = np.vstack((x,y)).T
    gradient = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
    
    print("****************************************************************")
    plt.clf()
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("y(x)")
    plt.show()
    plt.clf()
    plt.plot(x[1:], gradient)
    plt.xlabel("x")
    plt.ylabel("dy/dx")
    plt.title("Gradient dy/dx")
    plt.show()
    
    return array_xy, gradient

def function1(array_x):
    array_y = array_x ** 2
    return array_y

def function2(array_x):
    array_y = 2*(array_x**2) + 2**array_x
    return array_y

def function3(array_x):
    array_y = np.sin(array_x**(1/2))
    return array_y

array_xy, gradient = compute_gradient(function1)
array_xy, gradient = compute_gradient(function2)
array_xy, gradient = compute_gradient(function3, x_range=(0,50.1,0.1))
```

# [Problem 6] Find the minimum value

```python
array_xy, gradient = compute_gradient(function1)
min_index = np.argmin(array_xy[:,1])
print("y_min = y[{}] = {}\n".format(min_index, np.min(array_xy[:,1])))
print("Slope before minimum: gradient[{}] = {}".format(min_index-1, gradient[min_index-1]))
print("Slope after minimum: gradient[{}] = {}\n".format(min_index+1, gradient[min_index+1]))

array_xy, gradient = compute_gradient(function2)
min_index = np.argmin(array_xy[:,1])
print("y_min = y[{}] = {:.2f}\n".format(min_index, np.min(array_xy[:,1])))
print("Slope before minimum: gradient[{}] = {}".format(min_index-1, gradient[min_index-1]))
print("Slope after minimum: gradient[{}] = {}\n".format(min_index+1, gradient[min_index+1]))

array_xy, gradient = compute_gradient(function3, x_range=(0,50.1,0.1))
min_index = np.argmin(array_xy[:,1])
print("y_min = y[{}] = {:.2f}".format(min_index, np.min(array_xy[:,1])))
print("Slope before minimum: gradient[{}] = {}".format(min_index-1, gradient[min_index-1]))
print("Slope after minimum: gradient[{}] = {}\n".format(min_index+1, gradient[min_index+1]))
```
