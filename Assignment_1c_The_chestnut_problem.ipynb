# [Problem] The day when chestnut buns cover the solar system

```
def fill_up(small, large, small_name, large_name):
    n = 0
    volume = [small]
    while (volume[-1] < large):
        n += 1
        volume.append(small * 2**n)
    volume.append(large)
    print(volume)

    print("It takes {} minutes for the {} to fill up the entire {}.".format(5*n, small_name, large_name))
    plt.plot([5*i for i in range(n+1)], [volume[i]*100/large for i in range(n+1)], 'b-')
    plt.xlabel("Time (minutes)")
    plt.ylabel("Volume ratio (%)")
    plt.title("Number of minutes for the {} to fill up the entire {}.".format(small_name, large_name))
    plt.grid()
    

fill_up(5, 100000, "bun", "universe")
```
